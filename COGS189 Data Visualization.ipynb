```python
import numpy as np
import matplotlib as plt
import csv
import pandas as pd
```

```python
lst = []
with open("Kevin.csv",newline = '') as csvfile:
    eeg_kevin = csv.reader(csvfile, delimiter='\t',quotechar = '|')
    for i in eeg_kevin:
        lst.append(i)
kevin_df = pd.DataFrame(lst)
```
